# Apriori Algorithm

## Generating Frequent Itemsets

We map each item that CustomerID buys and store it into `dataset`.

```python
dataset = []
userBuy = {}

with open('market-basket.csv', 'r') as f:
    lines = f.readlines()
    firstRow = True
    n = 1
    for l in lines:
        if firstRow:
            firstRow = False
            continue
        data = l.split(';')
        if n < 5:
            print(data)
            n += 1
        user = data[5]
        val = data[1]

        if user not in userBuy.keys():
            userBuy[user] = [val]
        else:
            if val not in userBuy[user]:
                userBuy[user].append(val)

for key in userBuy:
    temp = [str(userBuy[key][0])]
    for i in range(1, len(userBuy[key])):
        temp.append(str(userBuy[key][i]))
    dataset.append(temp)
```

```python
print(dataset[1])
```

Transform data into the right format.

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

Return the items and itemsets with at least 60% support.

```python
from mlxtend.frequent_patterns import apriori

apriori(df, min_support=0.1)
```

 For better readability, we can set use_colnames=True.

```python
apriori(df, min_support=0.1, use_colnames=True)
```

## Selecting and Filtering Results

Let's assume we are only interested in itemsets of length 1 that have a support of at least 10 percent.

We create the frequent itemsets via apriori and add a new column that stores the length of each itemset.

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

We can select the results that satisfy our desired criteria as follows.

```python
frequent_itemsets[ (frequent_itemsets['length'] == 1) &
                   (frequent_itemsets['support'] >= 0.1) ]
```

Similarly, using the Pandas API, we can select entries based on the "itemsets" column.

```python
frequent_itemsets[ frequent_itemsets['itemsets'] == {'VINTAGE SNAP CARDS'} ]
```

## Working with Sparse Representations

To save memory, we represent your transaction data in the sparse format.

```python
oht_ary = te.fit(dataset).transform(dataset, sparse=True)
sparse_df = pd.DataFrame.sparse.from_spmatrix(oht_ary, columns=te.columns_)
sparse_df
```

```python
results = apriori(sparse_df, min_support=0.1, use_colnames=True, verbose=1)

results
```

Thus, `ASSORTED COLOUR BIRD ORNAMENT` is the most item that customer bought.

## Export to xlsx

```python
results_df = pd.DataFrame(results)
results_df.to_csv('results.csv', index=False)
```
