# Data Driven Dealings Development

```python
import pandas as pd
data = pd.read_csv('market-basket.csv', sep=";", dtype='unicode')
data.head()
```

```python
df1 = data[['CustomerID','Itemname']]
df1.head()
```

```python
df=df1.groupby ('CustomerID') ['Itemname']. apply (lambda x: x.reset_index (drop = True)). unstack (). reset_index ()
df.drop('CustomerID',axis=1, inplace=True)
df.head()
```

```python
items = (df[0].unique())
items
```

```python
encoded_vals = []
def custom():
    for index, row in df.iterrows():
        labels = {}
        uncommons = list(set(items) - set(row))
        commons = list(set(items).intersection(row))
        for uc in uncommons:
            labels[uc] = 0
        for com in commons:
            labels[com] = 1
        encoded_vals.append(labels)
custom()
ohe_df = pd.DataFrame(encoded_vals)
print(ohe_df)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
freq_items = apriori(ohe_df, min_support=0.1, use_colnames=True) #, verbose=1 #min_support=0.006, 
freq_items
```
